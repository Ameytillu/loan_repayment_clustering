In this step, I converted the categorical column purpose into numerical dummy variables using pd.get_dummies(). Since machine learning algorithms like KMeans cannot directly process text categories, encoding is necessary to convert them into a numeric format.

I used drop_first=True to avoid the dummy variable trap and reduce multicollinearity by dropping one of the categories. This keeps the dataset clean and ensures that all encoded variables represent unique information.

```python
df_encoded = pd.get_dummies(df, columns=['purpose'], drop_first=True)
df_encoded.head()
```

In this step, I prepared the dataset for the KMeans model by scaling all numerical features. I used StandardScaler from sklearn, which standardizes the values so that each feature has a mean of 0 and a standard deviation of 1.

Scaling is mportant before applying KMeans because the algorithm uses distance-based calculations. If the dataset contains features with different scales (for example, loan amount and interest rate), features with larger numerical ranges may dominate the clustering results. Standardization ensures that all features contribute equally to the model.

```python
from sklearn.preprocessing import StandardScaler
X = df_encoded.drop('not.fully.paid', axis=1)
y = df_encoded['not.fully.paid']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

In this step, I applied the KMeans clustering algorithm to group customers into two clusters. Since the dataset contains two repayment outcomes (fully paid vs. not fully paid), I set n_clusters=2.

```python
from sklearn.cluster import KMeans
from sklearn.metrics import confusion_matrix, classification_report, f1_score
kmeans = KMeans(n_clusters=2, random_state=42, n_init=20)
kmeans.fit(X_scaled)

df_encoded['cluster'] = kmeans.labels_

# Evaluate initial performance
print("Confusion Matrix:\n", confusion_matrix(y, df_encoded['cluster']))
print("\nClassification Report:\n", classification_report(y, df_encoded['cluster']))
print("F1 Score:", f1_score(y, df_encoded['cluster']))
```

In this step, I created F1-score function to manually calculate the precision, recall, and F1 value for the 1 class, which represents customers who did not fully pay their loans.

The confusion matrix shows that the model correctly identified 4330 customers who fully paid and 433 customers who did not fully pay their loans. However, it also made a significant number of misclassifications.

From the classification report, the precision for class 0 (fully paid) is 0.80, which means the model is fairly accurate when predicting customers who repaid their loans. In contrast, the precision for class 1 (not fully paid) is very low at 0.10, indicating that the model struggles to identify defaulters correctly.

The recall for class 1 is 0.28, showing that only 28% of actual defaulters were captured. The overall F1 score of 0.15 suggests weak performance in detecting customers who did not fully pay.

```python
# USER-DEFINED F1 SCORE FUNCTION 

from sklearn.metrics import confusion_matrix

def manual_f1_score(y_true, y_pred):
    """
    Custom F1-score function as required in the assignment.
    Calculates precision and recall for the '1' class (not fully paid).
    """
    cm = confusion_matrix(y_true, y_pred)
    tn, fp, fn, tp = cm.ravel()

    # Precision and Recall for class "1"
    precision = tp / (tp + fp) if (tp + fp) != 0 else 0
    recall = tp / (tp + fn) if (tp + fn) != 0 else 0

    # F1 Score
    if (precision + recall) == 0:
        return 0
    f1 = 2 * (precision * recall) / (precision + recall)
    return f1

manual_f1 = manual_f1_score(y, df_encoded['cluster'])
print("Manual F1 Score (for class 1 - not fully paid):", manual_f1)
```

- The F1 score for the 1 class (not fully paid) is approximately 0.15, it is low. This tells me that the KMeans clustering model is not very effective at identifying customers who did not fully repay their loans.

- A low F1 score indicates two things:

Low precision: Many of the customers predicted as not fully paid actually belong to the fully paid group. This means the model produces a high number of false positives.

Low recall: The model is also missing a large portion of the actual not fully paid customers, resulting in a high number of false negatives.

- Since KMeans is an unsupervised algorithm, it does not learn directly from the repayment labels. Instead, it groups customers based only on similarities in the feature space. Because the not fully paid group is much smaller and overlaps heavily with the fully paid group, the model struggles to form a clean boundary around them.
